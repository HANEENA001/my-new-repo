```python
df.head(5)
```

## Step 4: Data Analysis - perform basic data analysis tasks:

### a) Total revenue for each product

```python
total_revenue = df.groupby('Product')['Revenue'].sum()
print(total_revenue)
```

### b) Total units sold for each product

```python
units_sold = df.groupby('Product')['Order_Quantity'].sum()
print(units_sold)
```

### c) Monthly revenue trends

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

monthly_rev = df.groupby(['Year', 'Month'])['Revenue'].sum()
print(monthly_rev)
```

### d) Average revenue per unit for each product

```python
average_rev = df.groupby('Product')['Revenue'].mean() / df.groupby('Product')['Order_Quantity'].mean()
print(average_rev)
```

## Step 5: Data Visualization

```python
plt.figure(figsize=(15, 6))
monthly_rev.plot(kind='line', marker='o')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Data Visualization of Monthly Revenue')
plt.show()
```
