# 4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
not_null_details = df_new[~df_new['crew'].isna()]
```

```python
not_null_details.shape
```

```python
def producers_tolist(index):
  movie_index = not_null_details.iloc[index]
  details_tolist= movie_index.loc['crew']
  producer_tolist=[]
  for elem in details_tolist:
     if elem['job']=='Producer':
        producer_tolist.append(elem['name'])
  return producer_tolist
```

```python
producers_tolist(61)
```

```python
def directors_tolist(index):
  index = not_null_details.iloc[index]
  details_tolist= index.loc['crew']

  for elem in details_tolist:
     if elem['job']=='Director':
        return elem['name']
```

```python
directors_tolist(61)
```

```python
totable = pd.DataFrame(columns=['Movie Title','Producers','Directors','ROI'])
```

```python
for index,row in not_null_details.iterrows():

  try:
      totable = totable.append({'Movie Title':not_null_details.loc[index,'original_title'],'Producers':producers_tolist(index),'Directors':directors_tolist(index),'ROI':not_null_details.loc[index,'roi']},ignore_index=True)
  except:
    continue
```

# 5. Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
actors_tolist=[]
for index,row in not_null_cast.iterrows():
  for iter in not_null_cast.loc[index,'cast']:
    if type(iter)== dict:
      actor= iter['name']
      actors_tolist.append(actor)
```

```python
tolist_actors.shape
```

```python
tolist_actors.head()
```

```python
tolist_actors.value_counts().reset_index().head()
```

```python
max_pro_1=[]
max_pro_2=[]
pro_movie_1=[]
pro_movie_2=[]
for index,row in not_null_cast.iterrows():
  for iter in not_null_cast.loc[index,'cast']:
    if type(iter)== dict:
      actor= iter['name']
      if 'Robert De Niro' in actor:
        max_pro_1.append(not_null_cast.loc[index,'profit'])
        pro_movie_1.append(not_null_cast.loc[index,'original_title'])
      if 'Samuel L. Jackson' in actor:
        max_pro_2.append(not_null_cast.loc[index,'profit'])
        pro_movie_2.append(not_null_cast.loc[index,'original_title'])
```

```python
gener_tolist=[]
a=[]
for i in range(len(pro_movie_1)):
  for g in not_null_cast.loc[i,'genres']:
    a.append(g['name'])

  gener_tolist.append(a)
  a=[]

generes_tolist=[]
b=[]
for i in range(len(pro_movie_2)):
  for g in not_null_cast.loc[i,'genres']:
    b.append(g['name'])

  generes_tolist.append(b)
  b=[]
```

```python
gener = np.array(gener_tolist)
generes = np.array(generes_tolist)
```

```python
movie_no_1= pd.DataFrame(pro_movie_1,columns=['Movie Name'])
profite_no_1=pd.DataFrame(max_pro_1,columns=['Movie Profit'])
gener_no_1= pd.DataFrame(gener.flatten(),columns=['Genres'])
```

```python
Movies_by_Robert=pd.concat([movie_no_1,gener_no_1,profite_no_1],axis=1)
```

```python
Movies_by_Robert.sort_values(by='Movie Profit',ascending=False).head()
```

```python
movie_no_2= pd.DataFrame(pro_movie_2,columns=['Movie Name'])
profite_no_2=pd.DataFrame(max_pro_2,columns=['Movie Profit'])
gener_no_2= pd.DataFrame(generes.flatten(),columns=['Genres'])
```
