# <u>**Problem 1**</u>

### Plot the total profit across the months

* X label name = Month Number
* Y label name = Total profit



```
profit = df ['total_profit'].tolist()
month = df ['month_number'].tolist()
plt.plot(month, profit, label = 'Total profit across the months')
plt.xlabel('month_number')
plt.ylabel('total_Profit')
plt.xticks(month)
plt.title('Total profit across the months')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
month = df ['month_number'].tolist()
faceCrem = df ['facecream'].tolist()
faceWash = df ['facewash'].tolist()

plt.bar([a-0.25 for a in month], faceCrem, width= 0.25, label = 'Face Cream ',align='edge')
plt.bar([a+0.25 for a in month], faceWash, width= -0.25, label = 'Face Wash ', align='edge')

plt.xlabel('month_number')
plt.ylabel('total_units')
plt.title('Face Cream and Facewash')
plt.legend(loc='upper left')
plt.title('Sales data')

plt.xticks(month)
plt.title('Facewash and facecream')
plt.show()
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
month_number = df["month_number"].tolist()
facecream = df["facecream"].tolist()
facewash = df["facewash"].tolist()
toothpaste = df["toothpaste"].tolist()
bathingsoap = df["bathingsoap"].tolist()
shampoo = df["shampoo"].tolist()
moisturizer = df["moisturizer"].tolist()

plt.plot(month_number, facecream, marker="*", label="Face Cream",linewidth="3")
plt.plot(month_number, facewash, marker="*",label="Facewash", linewidth="6")
plt.plot(month_number, toothpaste, marker="*",label="toothpaste",linewidth="3")
plt.plot(month_number, bathingsoap, marker="*", label="bathingsoap", linewidth="3")
plt.plot(month_number, shampoo,marker="*", label="shampoo", linewidth="3")
plt.plot(month_number, moisturizer, marker="*", label="moisturizer ", linewidth="3")

plt.title("Units Sold Per Month")
plt.xlabel("month_number")
plt.ylabel("total_profit")
plt.legend()
plt.show()
```
